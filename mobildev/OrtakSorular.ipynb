```

LCSRe(i, j)i ve j karakterleriyle biten eşleşen ve örtüşmeyen alt dizelerin uzunluğunu saklar.

If str[i-1] == str[j-1] && (j-i) > LCSRe(i-1, j-1)
     LCSRe(i, j) = LCSRe(i-1, j-1) + 1, 
Else
     LCSRe(i, j) = 0
i'nin 1'den n'ye değiştiği ve j'nin i+1'den n'ye değiştiği yerde
```



````python

def longestRepeatedSubstring(str):
 
    n = len(str)
    LCSRe = [[0 for x in range(n + 1)]
                for y in range(n + 1)]
 
    res = "" 
    res_length = 0 

    index = 0
    for i in range(1, n + 1):
        for j in range(i + 1, n + 1):
             

            if (str[i - 1] == str[j - 1] and
                LCSRe[i - 1][j - 1] < (j - i)):
                LCSRe[i][j] = LCSRe[i - 1][j - 1] + 1
 

                if (LCSRe[i][j] > res_length):
                    res_length = LCSRe[i][j]
                    index = max(i, index)
                 
            else:
                LCSRe[i][j] = 0
 

    if (res_length > 0):
        for i in range(index - res_length + 1,
                                    index + 1):
            res = res + str[i - 1]
 
    return res
 
if __name__ == "__main__":
    A = [1, 70, 9, 1, 2, 30, 6, 1, 2, 30, 50 ]
    output = [str(x)+" " for x in A]
    result=longestRepeatedSubstring(output)
    
    
    
    print(result)
 
````

a-)Bilgisayarla görme dersinde hocamız output size hesaplarken verilen formül sonucunun tam sayı olması gerektiğini dile getirmişti.İşlemler sırasında 5x5 input 2x2 fitre stride 2 ve zero padding sonucu bu değer ondalıklı çıkmakta.Araştırıp kesin bir sonuç elde edemedim.

mnist.png

````python
import numpy as np
from keras.utils.np_utils import to_categorical 
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D,AvgPool2D
import tensorflow as tf


(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

print(x_train.shape) # (60000, 5, 5)
print(x_test.shape) # (60000, 5, 5)

# train set / data 
x_train = np.expand_dims(x_train, axis=-1)
x_test = np.expand_dims(x_test, axis=-1)

x_train = tf.image.resize(x_train, [5,5])
x_test = tf.image.resize(x_test, [5,5])

x_train = x_train / 255
x_test = x_test / 255

print(x_train.shape) # (60000, 5, 5, 1)
print(x_test.shape) # (60000, 5, 5, 1)





model = Sequential()
model.add(Conv2D(filters = 2, kernel_size = (2,2),padding = 'Same', activation ='relu', input_shape = (5,5,1)))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))

model.add(Conv2D(filters = 2, kernel_size = (2,2),padding = 'Same', activation ='relu'))
model.add(AvgPool2D(pool_size=(2,2), strides=(2,2)))

model.add(Flatten())
model.add(Dense(10, activation = "softmax"))


model.summary()



model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(x_train, y_train,
          batch_size=128, epochs=100,
          verbose=1)







````
